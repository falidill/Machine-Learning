Scatter plot of 'charges' versus 'bmi' :

1. X-axis (bmi):This represents the body mass index of individuals in the dataset. BMI is a common measure to understand whether an individual has 
a healthy body weight for their height.

2. Y-axis (charges): This  represents the insurance charges or costs associated with each individual.

3. Data Distribution:*
   - The scatter of points mostly along the lower part of the plot suggests that for a wide range of BMIs, the insurance charges are on the lower end.
   - There is a significant number of points that spread upwards, particularly as BMI increases. This might indicate that higher BMIs could be 
   associated with higher insurance charges, which is plausible if we consider that higher BMI might be associated with higher health risks.
   - There's a notable separation between two clusters of data points vertically. This could suggest that there are two distinct groups in the dataset,
   possibly related to an underlying factor like whether an individual is a smoker or non-smoker, as this is a common factor influencing insurance costs.

4. Potential Outliers:There seem to be no clear outliers on the lower end of the charges, but at higher BMIs, the charges vary greatly, which could 
   indicate some outliers or a non-linear relationship.

5. Overall Trend: There doesn't seem to be a clear linear trend across the entire range of BMI values. However, there may be a trend where higher BMI 
 is associated with a wider range of charges and potentially higher charges on average. A more detailed analysis, possibly including a line of best fit 
 or a regression analysis, could give more insights into this trend.

```python
import scipy.stats as stats

# T-test: Charges by smoker status
charges_smoker_yes = df[df['smoker'] == 'yes']['charges']
charges_smoker_no = df[df['smoker'] == 'no']['charges']

t_stat, p_val = stats.ttest_ind(charges_smoker_yes, charges_smoker_no)

print(f"T-test for Smoker vs. Non-smoker: t-stat={t_stat:.2f}, p-val={p_val:.5f}")

# ANOVA: Charges by region
charges_nw = df[df['region'] == 'northwest']['charges']
charges_ne = df[df['region'] == 'northeast']['charges']
charges_se = df[df['region'] == 'southeast']['charges']
charges_sw = df[df['region'] == 'southwest']['charges']

f_stat, p_val = stats.f_oneway(charges_nw, charges_ne, charges_se, charges_sw)

print(f"ANOVA for Regions: F-stat={f_stat:.2f}, p-val={p_val:.5f}")
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['smoker'], df['bmi_category'])

chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Squared Test: chi2={chi2_stat:.2f}, p-val={p_val:.5f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['bmi'])
plt.title('Box Plot of BMI')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['charges'])
plt.title('Box Plot of Charges')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['age'])
plt.title('Box Plot of Age')

plt.tight_layout()
plt.show()
```

```python
Q1_bmi = df['bmi'].quantile(0.25)
Q3_bmi = df['bmi'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi

Q1_charges = df['charges'].quantile(0.25)
Q3_charges = df['charges'].quantile(0.75)
IQR_charges = Q3_charges - Q1_charges

Q1_age = df['age'].quantile(0.25)
Q3_age = df['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

lower_bmi = Q1_bmi - 1.5 * IQR_bmi
upper_bmi = Q3_bmi + 1.5 * IQR_bmi

lower_charges = Q1_charges - 1.5 * IQR_charges
upper_charges = Q3_charges + 1.5 * IQR_charges

lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age

outliers_bmi = df[(df['bmi'] < lower_bmi) | (df['bmi'] > upper_bmi)]
outliers_charges = df[(df['charges'] < lower_charges) | (df['charges'] > upper_charges)]
outliers_age = df[(df['age'] < lower_age) | (df['age'] > upper_age)]

print("Number of outliers in 'bmi':", len(outliers_bmi))
print("Number of outliers in 'charges':", len(outliers_charges))
print("Number of outliers in 'age':", len(outliers_age))
```
